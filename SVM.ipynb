```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("Petrol.csv")
data = data.replace({"January":1,"February":2,"March":3,"April":4,"May":5,"June":6,"July":7,"August":8,"September":9,"October":10,"November":11,"December":12})
```

```python
data
```

```python
X = data[["Year","Month"]]
y = data[["Petrol Price (PKR)"]]
scaler = StandardScaler()
X_Scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_Scaled, y, test_size=0.2, random_state=42)

svr_model = SVR(kernel='rbf', C=50, epsilon=0.1)
svr_model.fit(X_train, y_train)

y_pred = svr_model.predict(X_test)

r2 = r2_score(y_test, y_pred)
print("R2 Score:", r2)
```

```python
new_data = pd.DataFrame({'Year': [2090], 'Month': [12]})
new_data_scaled = scaler.transform(new_data)
predicted_price = svr_model.predict(new_data_scaled)
print("Predicted Petrol Price:", predicted_price[0])
```
