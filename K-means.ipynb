```python

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python

data = {
    'Living Thing': ["Human","Chimpanzee","Orangutan","Gorilla","Dolphin","Elephant","Pig","Dog","Cat","Octopus"],
    'Estimated IQ': ["100","20","25","28","50","40","32","26","18","45"],
}
data = pd.DataFrame(data)
data
```

```python
# Prepare the feature for KMeans
X = data['Estimated IQ'].values.reshape(-1, 1)  # Reshape to 2D array
```

```python

kmean = KMeans(n_clusters=3, random_state=42).fit(X)

data['Cluster'] = kmean.labels_

data
```

```python
# Visualize clusters
plt.figure(figsize=(6, 4))
for cluster in range(3):
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(cluster_data['Estimated IQ'], [cluster] * len(cluster_data), label=f'Cluster {cluster}')
```
